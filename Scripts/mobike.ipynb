```python
df_criterias = pd.read_pickle('C:/Users/jeane/Documents/Travail/UTSEUS/Urban Data Hackathon/pickles/meters/communities-POI_Subway_taxi-meters.pk')
df_criterias = gpd.GeoDataFrame(df_criterias, geometry = [Point(float(x),float(y)) for x,y in zip(df_criterias.transformed_longitude, df_criterias.transformed_latitude)])
dist_varation = {50, 100, 200, 400}
df_criterias.head()
```

```python
%%time
global i
i = 0

df = pd.read_pickle("C:/Users/jeane/Documents/Travail/UTSEUS/Urban Data Hackathon/pickles/meters/taxi_meters.pk")
df = gpd.GeoDataFrame(df, geometry = [Point(float(x),float(y)) for x,y in zip(df.transformed_longitude, df.transformed_latitude)])


for dist in dist_varation :
    df_criterias["mobikeDeparture" + str(dist)] = pd.Series([0 for x in range(len(df_criterias.index))])
    nbMobike = []

    for j, poi in df_criterias.iterrows():

        gdf_f = df[df.geometry.within(poi.geometry.buffer(400))]
        nbMobike.append(len(gdf_f))

        i += 1
        clear_output(wait = True)
        print(i, "/", len(df_criterias)*len(dist_varation))

    df_criterias["mobikeDeparture" + str(dist)] = nbMobike
```

```python
df_criterias.to_pickle('C:/Users/jeane/Documents/Travail/UTSEUS/Urban Data Hackathon/pickles/meters/communities-POI_Subway_taxi_mobike-meters.pk')
```
