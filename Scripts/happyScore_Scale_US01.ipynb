```python
df = pd.read_pickle('C:/Users/jeane/Documents/Travail/UTSEUS/Urban Data Hackathon/pickles/meters/communities-POI_Subway_taxi-meters.pk')
df_criterias = df
```

```python
df_criterias.head()
```

```python
def createHappyScore(clean, smell, noise):
    happyScore = ""
    if clean < 3:
        happyScore += "Dirty"
    else : 
        happyScore += "Clean"
    
    if smell == 1:
        happyScore += "Notsmelly"
    else : 
        happyScore += "Smelly"
        
    if noise == 1:
        happyScore += "Quite"
    else:
        happyScore += "Noisy"
    
    return happyScore
    
df_criterias["hapiness"] = df_criterias.apply(lambda x: createHappyScore(x.clean, x.smell, x.noise), axis=1)
```

```python
df_criterias.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
nameColumn = []

#X2['restaurant']=scaler.fit_transform(X[['restaurant']])

for column in df_criterias:
    nameColumn.append(column)
    
for nameC in nameColumn:
    if nameC != "clean" and nameC != "smell" and nameC != "noise" and nameC != "hapiness":
        df_criterias[nameC]=scaler.fit_transform(df_criterias[[nameC]])
    
```
