```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/jeane/Desktop/USXX/UTSEUS-anjuke-real-estate.csv', index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='surface', y='price', data=df)
```

```python
sns.lmplot(x='surface', y='price', data=df,
           fit_reg=False, # No regression line
           hue='district')   # Color by evolution stage

plt.yscale('log')
plt.xscale('log')
```

```python
sns.boxplot(data=df)
```

```python
# Pre-format DataFrame
stats_df = df.drop(['longitude', 'latitude', 'onesquaremeter', 'price', 'done', 'surface'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')
 
# Violin plot
sns.violinplot(x='district', y='surface', data=df)
```

```python
sns.set_style('whitegrid')
 
# Violin plot
sns.violinplot(x='district', y='price', data=df)
```
