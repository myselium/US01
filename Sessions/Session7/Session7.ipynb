```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/jeane/Desktop/USXX/data-science-course/UTSEUS-MOBIKE-shanghai_full.csv', index_col=0)
```

```python
df.head()
```

```python
%matplotlib inline
df.plot.scatter(x="start_location_x", y="start_location_y");
```

```python
Ymin = 30.975;
Xmin = 121.10157;
Ymax = 31.514999;
Xmax = 121.804611;

df = df[(df["start_location_y"] <= Ymax) & (df["start_location_y"] >= Ymin) & (df["start_location_x"] <= Xmax) & (df["start_location_x"] >= Xmin) & (df["end_location_y"] <= Ymax) & (df["end_location_y"] >= Ymin) & (df["end_location_x"] <= Xmax) & (df["end_location_x"] >= Xmin)]
```

```python
%matplotlib inline
df.plot.scatter(x="start_location_x", y="start_location_y", s=0.005, c="#58D3F7", figsize=(20,15));
```

```python
df["start_time_timestamp"] = df["start_time"].apply(lambda x: pd.Timestamp(x))
df["end_time_timestamp"] = df["end_time"].apply(lambda x: pd.Timestamp(x))
df.head()
```
